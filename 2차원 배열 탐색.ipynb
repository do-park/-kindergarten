```python
N, M = map(int, input().split())
area = [list(map(int, input().split())) for _ in range(N)]
result = 0

# 정사각형 테트로미노
for i in range(0, N-1):
    for j in range(0, M-1):
        sum = area[i][j] + area[i+1][j] + area[i][j+1] + area[i+1][j+1]
        if result < sum:
            result = sum

# 직사각형 테트리미노 가로
for i in range(0, N):
    for j in range(0, M-3):
        sum = area[i][j] + area[i][j+1] + area[i][j+2] + area[i][j+3]
        if result < sum:
            result = sum
# 직사각형 테트리미노 세로
for i in range(0, N-3):
    for j in range(0, M):
        sum = area[i][j] + area[i+1][j] + area[i+2][j] + area[i+3][j]
        if result < sum:
            result = sum

# ㄴ 테트리미노 1 (세로가 긴 ㄴ)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j] + area[i+1][j] + area[i+2][j] + area[i+2][j+1]
        if result < sum:
            result = sum

# ㄴ 테트리미노 2 (세로가 긴 ㄴ 대칭)
for i in range(0, N-2):
    for j in range(1, M):
        sum = area[i][j] + area[i+1][j] + area[i+2][j] + area[i+2][j-1]
        if result < sum:
            result = sum

# ㄴ 테트리미노 3 (세로가 긴 ㄱ)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j] + area[i][j+1] + area[i+1][j+1] + area[i+2][j+1]
        if result < sum:
            result = sum

# ㄴ 테트리미노 4 (세로가 긴 ㄱ 대칭)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j] + area[i][j+1] + area[i+1][j] + area[i+2][j]
        if result < sum:
            result = sum

# ㄴ 테트리미노 5 (가로가 긴 ㄱ)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j] + area[i][j+1] + area[i][j+2] + area[i+1][j+2]
        if result < sum:
            result = sum

# ㄴ 테트리미노 6 (가로가 긴 ㄱ 대칭)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j] + area[i][j+1] + area[i][j+2] + area[i+1][j]
        if result < sum:
            result = sum

# ㄴ 테트리미노 7 (가로가 긴 ㄴ)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j] + area[i+1][j] + area[i+1][j+1] + area[i+1][j+2]
        if result < sum:
            result = sum

# ㄴ 테트리미노 7 (가로가 긴 ㄴ 대칭)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j+2] + area[i+1][j] + area[i+1][j+1] + area[i+1][j+2]
        if result < sum:
            result = sum

# ㅗ 테트리미노 1 (ㅗ)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j+1] + area[i+1][j] + area[i+1][j+1] + area[i+1][j+2]
        if result < sum:
            result = sum

# ㅗ 테트리미노 2 (ㅜ)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j] + area[i][j+1] + area[i][j+2] + area[i+1][j+1]
        if result < sum:
            result = sum


# ㅗ 테트리미노 3 (ㅓ)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j+1] + area[i+1][j] + area[i+1][j+1] + area[i+2][j+1]
        if result < sum:
            result = sum


# ㅗ 테트리미노 4 (ㅏ)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j] + area[i+1][j] + area[i+1][j+1] + area[i+2][j]
        if result < sum:
            result = sum

# 마지막 테트리미노 1 (ㄴㄱ 세로)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i][j] + area[i+1][j] + area[i+1][j+1] + area[i+2][j+1]
        if result < sum:
            result = sum

# 마지막 테트리미노 2 (ㄱㄴ 가로)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j] + area[i][j+1] + area[i+1][j+1] + area[i+1][j+2]
        if result < sum:
            result = sum
# 마지막 테트리미노 3 (1 대칭)
for i in range(0, N-2):
    for j in range(0, M-1):
        sum = area[i+1][j] + area[i+2][j] + area[i][j+1] + area[i+1][j+1]
        if result < sum:
            result = sum
# 마지막 테트리미노 4 (2 대칭)
for i in range(0, N-1):
    for j in range(0, M-2):
        sum = area[i][j+1] + area[i][j+2] + area[i+1][j] + area[i+1][j+1]
        if result < sum:
            result = sum

print(result)
```

```python
A, B = map(int, input().split())
board = [list(input()) for _ in range(0, A)]
chess = [[0] * B for _ in range(0, A)]

# chess: 검흰검흰...
for a in range(0, A):
    for b in range(0, B):
        if a % 2 == 0:
            if b % 2 == 0:
                if board[a][b] != 'B':
                    chess[a][b] = 1
            else:
                if board[a][b] != 'W':
                    chess[a][b] = 1
        else:
            if b % 2 == 0:
                if board[a][b] != 'W':
                    chess[a][b] = 1
            else:
                if board[a][b] != 'B':
                    chess[a][b] = 1

count = 0
counts = []

for a in range(0, A - 7):
    for b in range(0, B - 7):
        for i in range(0, 8):
            for j in range(0, 8):
                if chess[a + i][b + j] == 1:
                    count += 1
        counts.append(count)
        count = 0

print('{}'.format(min(min(counts), 64 - max(counts))))
```

```python
def bomboni():
    for y in range(N):
        temp = 1
        for x in range(1, N):
            if board[y][x - 1] == board[y][x]:
                temp += 1
            else:
                ans.append(temp)
                temp = 1
        ans.append(temp)
    ans.append(temp)
    for x in range(N):
        temp = 1
        for y in range(1, N):
            if board[y - 1][x] == board[y][x]:
                temp += 1
            else:
                ans.append(temp)
                temp = 1
        ans.append(temp)
    ans.append(temp)


N = int(input())
board = [list(map(str, input())) for _ in range(N)]
ans = []
for i in range(1, N):
    for j in range(1, N):
        board[i][j - 1], board[i][j] = board[i][j], board[i][j - 1]
        bomboni()
        board[i][j - 1], board[i][j] = board[i][j], board[i][j - 1]
        board[i - 1][j], board[i][j] = board[i][j], board[i - 1][j]
        bomboni()
        board[i - 1][j], board[i][j] = board[i][j], board[i - 1][j]

print(max(ans))
```
